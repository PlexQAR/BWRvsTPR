```python
import numpy as np
import matplotlib.pyplot as plt
import openmc

# Load statepoint file (from a previous run)
sp = openmc.StatePoint('statepoint.100.h5')
```

```python
# Access the tallies
neutron_flux = sp.get_tally(name='neutron_flux')
power_distribution = sp.get_tally(name='power_distribution')
```

```python
# Get the results
flux_results = neutron_flux.mean
power_results = power_distribution.mean
```

```python
mesh_dims = [100, 100, 100]

flux_reshaped = flux_results.reshape(mesh_dims)
power_reshaped = power_results.reshape(mesh_dims)

z_slice = 50
```

```python
# Plot the neutron flux for the selected z slice
plt.figure(figsize=(8, 6))
plt.imshow(flux_reshaped[:, :, z_slice], origin='lower', extent=[-50, 50, -50, 50])
plt.colorbar(label='Neutron Flux')
plt.title(f'Neutron Flux at Z = {z_slice}')
plt.xlabel('X Position [cm]')
plt.ylabel('Y Position [cm]')
plt.show()

# Plot the power distribution for the selected z slice
plt.figure(figsize=(8, 6))
plt.imshow(power_reshaped[:, :, z_slice], origin='lower', extent=[-50, 50, -50, 50])
plt.colorbar(label='Power Distribution')
plt.title(f'Power Distribution at Z = {z_slice}')
plt.xlabel('X Position [cm]')
plt.ylabel('Y Position [cm]')
plt.show()
```
