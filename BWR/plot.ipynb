```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the plot data
with h5py.File('plot_7.h5', 'r') as f:
    data = f['/data'][:]  # Adjust the dataset path as necessary

# Create a figure for the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate grid coordinates
x, y, z = np.indices(data.shape)

# Set the threshold value for visualization
threshold = 0.5  # Adjust this based on your data

# Create a 3D scatter plot for non-zero values
ax.scatter(x[data > threshold], y[data > threshold], z[data > threshold], c='cyan', marker='o', alpha=0.1)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Voxel Plot')
plt.show()
```
